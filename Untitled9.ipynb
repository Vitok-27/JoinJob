#2 задача
##2.1. Изучите распределение домов от наличия вида на набережную.


```python
water=house['waterfront'].value_counts().reset_index()
water

```

```python
names= water.index
values=water.values
names
```

```python
plt.figure(figsize=(8, 4))
plt.pie(values, autopct='%1.1f%%',labels=names)
plt.legend(['no', 'yes']);
```

###Вывод: количество домов с видом на набережную составляет всего 0,8

##2.2. Изучите распределение этажей домов

```python
Floors=house['floors'].value_counts()
Floors
```

```python
names=Floors.index
values=Floors.values
```

```python
plt.figure(figsize=(8, 4))
plt.title('Количество этажей')
plt.pie(values, autopct='%1.1f%%',labels=names)
plt.legend(['no', 'yes']);
```

###Вывод: большую долю недвижимости составляют одноэтажные и духэтажные сооружения, 49.4% и 38.1% соответственно

##2.3. Изучите распределение состояния домов

```python
condition=house['condition'].value_counts()
condition
```

```python
names=condition.index
values=condition.values
```

```python
plt.figure(figsize=(8,4))
plt.title('Cостояние домов')
plt.pie(condition,autopct='%1.1f%%',labels=names)
plt.legend(['no', 'yes']);
```

###Вывод: 64.9% домов находятся в хорошом состоянии.
